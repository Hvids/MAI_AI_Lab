```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.simplefilter('ignore')
%matplotlib inline
```

```python
path1 = '.data/AADR.csv'
df = pd.read_csv(path1)
```

# Начальные сведения

```python
df.head()
```

```python
print(f'Размер {df.shape}')
print(f'Все колонки: {" ".join(df.columns.values)}')

number_columns = ['Open','High','Low',]
predict_column = 'Close'
date_column = 'Date'
deleted_columns = ['Adj Close']
df = df.drop(deleted_columns,axis = 1 )
print(f'Иследуемое значение:  {predict_column}')
print(f'Количественные признаки: {" ".join(number_columns)}')
print(f'Поле даты: {date_column}')
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
fig, ax = plt.subplots(ncols=1, nrows = 1, figsize= (12,12))
df.hist(bins = 50,ax=ax)

fig.savefig('./AADR_fig/hist.png')
plt.show()
```

# Преобразование Date в datetime

```python
df.Date = pd.to_datetime(df.Date)
```

# Добавление новой переменной

```python
df['t'] = df.index
```

# Ковариация

```python
corr_matrix = df.corr()
corr_matrix[predict_column].sort_values(ascending=False)
```

# Графики

```python
fig, ax = plt.subplots(ncols=1, nrows = 1, figsize= (12,12))
ax.set_title(f'Плотность распределения {predict_column}')
df[predict_column].plot.kde(ax=ax)
fig.savefig('./AADR_fig/kde.png')
plt.show()
```

```python
plot_arr =  number_columns
nrows = 3
ncols = 2

fig, axs = plt.subplots(ncols= ncols, nrows = nrows, figsize= (20,20))
fig.suptitle(f'Зависимость {predict_column} от признаков числовых ', size = 20)
plt.axis('off') 
for i in range(nrows):
    for j in range(ncols):
        if i+j >= len(plot_arr):
            continue
        column = plot_arr[i+j]        
        df.plot.scatter(x=column, y=predict_column, ax = axs[i][j])
fig.savefig('./AADR_fig/depend_number.png')
plt.show()
```

```python
fig, ax = plt.subplots(ncols=1, nrows = 1, figsize= (12,12))
ax.set_title(f'Зависимость {predict_column} от {date_column}')
df.plot(x = "Date", y = predict_column, ax = ax)
fig.savefig('./AADR_fig/depend_date.png')
plt.show()
```
