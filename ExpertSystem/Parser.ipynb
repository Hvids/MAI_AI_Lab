Формат базы знаний: предикат film(название фильма, режиссер, год, жанры, длительность, контент цензор, актер первого плана, актер второго плана, рейтинг imdb, ссылка на фильм на сайт imdb)

```python
'''film('Avatar ', 'James Cameron', 2009, ['Action', 'Adventure', 'Fantasy'], 'Sci-Fi', 178, 'PG-13', 'CCH Pounder', 'Joel David Moore', 7.9, 'http://www.imdb.com/title/tt0499549/?ref_=fn_tt_tt_1').'''

def splitter(str_):
    out = ''
    for item in str_.split('\''):
        out += item

    return out


def good_str(row):
    out = 'film('
    columns = []
    columns.append('movie_title')
    columns.append('director_name')
    columns.append('title_year')
    columns.append('genres')
    columns.append('duration')
    columns.append('content_rating')
    columns.append('actor_1_name')
    columns.append('actor_2_name')
    columns.append('imdb_score')
    columns.append('movie_imdb_link')
    
    for i in columns:
        if i == 'title_year' or i == 'duration':
            out += str(int(row[i])) + ', '
            
        elif i == 'genres':
            out += '['
            for item in row[i].split('|'):
                out += quotes(item) + ', '
            
            out = out[0:len(out) - 2]
            out += '], '
            
        elif i == 'imdb_score':
            out += str(row[i]) + ', '
            
        elif i == 'movie_imdb_link':
            out += quotes(row[i]) + ', '
            
        else:   
            out += quotes(splitter(row[i])) + ', '
        
    out = out[0:len(out) - 2]
    out += ').\n'
    return out

def write_actors(top):
    out = 'top_actors(['
    for each in top:
        out += quotes(each) + ', '
    out = out[0:len(out) - 2]
    out += ']).\n'
    
    return out

file.write(write_actors(top_20_actors))
file.write('\n')

for index, row in df.iterrows():    

    file.write(good_str(row))
file.close()
```
